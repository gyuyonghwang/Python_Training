```python
x = [0,1,2,3]
y = [0,2,4,6]

plt.title("Our first graph", fontdict={"fontname": "Comic Sans MS", "fontsize":20})
plt.xlabel("X Axis", fontdict={"fontname": "Comic Sans MS", "fontsize":12})
plt.ylabel("Y Axis")

plt.xticks([0,1,2,3])
plt.yticks([0,2,4,6])

plt.plot(x,y, label="2x", color="red", linewidth=2, marker =".", markersize=10, markeredgecolor="blue", linestyle="--") #label="2x" works together with plt.legend()
plt.legend()

plt.show(x,y)
```

```python
plt.title("Our first graph", fontdict={"fontname": "Comic Sans MS", "fontsize":20})
plt.xlabel("X Axis", fontdict={"fontname": "Comic Sans MS", "fontsize":12})
plt.ylabel("Y Axis")

# First line
plt.plot(x,y, label="2x", color="red", linewidth=2, marker =".", markersize=10, markeredgecolor="blue", linestyle="--") #label="2x" works together with plt.legend()


# Adding a second line
x2 = np.arange(0,4,0.5)
plt.plot(x2, x2**2)

plt.xticks([0,1,2,3])
plt.yticks([0,2,4,6])

plt.legend()

plt.show(x,y)
```

```python
# Resize the graph
plt.figure(figsize=(5,3), dpi=300)

plt.title("Our first graph", fontdict={"fontname": "Comic Sans MS", "fontsize":20})
plt.xlabel("X Axis", fontdict={"fontname": "Comic Sans MS", "fontsize":12})
plt.ylabel("Y Axis")

# First line
plt.plot(x,y, label="2x", color="red", linewidth=2, marker =".", markersize=10, markeredgecolor="blue", linestyle="--") #label="2x" works together with plt.legend()

# Adding a second line - first half
x2 = np.arange(0,4,0.5)
plt.plot(x2[:5],x2[:5]**2, marker =".", markersize=10, markeredgecolor="green") # Plot only first 4 values
# Adding a second line - second half
plt.plot(x2[4:],x2[4:]**2, marker =".", markersize=10, markeredgecolor="green", linestyle="--") # Plot the rest in cashed line


plt.xticks([0,1,2,3])
plt.yticks([0,2,4,6])

plt.legend()

plt.savefig("mygraph.png", dpi=300) #save graph, DPI is not a must parameter.

plt.show(x,y)
```

```python
plt.figure(figsize=(5,3), dpi=200)

plt.title("Gas prices over time in us dollars")
plt.xlabel("Year")
plt.ylabel("Gas Price in USD")
plt.xticks(gas["Year"][::3], rotation = 45) #[::3] means plot every 3 years

plt.plot(gas['Year'], gas['USA'], label="USA", marker =".", markersize=10)
plt.plot(gas['Year'], gas['Canada'], label="Canada", marker =".", markersize=10)
plt.plot(gas['Year'], gas['South Korea'], label="South Korea", marker =".", markersize=10)
plt.plot(gas['Year'], gas['France'], label="France", marker =".", markersize=10)



plt.legend()
plt.show()
```

```python
plt.figure(figsize=(5,3), dpi=200)

plt.title("Gas prices over time in us dollars")
plt.xlabel("Year")
plt.ylabel("Gas Price in USD")
plt.xticks(gas["Year"][::3], rotation = 45) #[::3] means plot every 3 years

# Using the for loop
for country in gas:
    if country != "Year":
        plt.plot(gas["Year"], gas[country],label=country, marker =".", markersize=10)


plt.legend(bbox_to_anchor=(1.40, 1.03), loc='upper right')

plt.savefig("Gas_price_figure.png")
plt.show()
```

# Pie Chart 1

```python
left = fifa.loc[fifa["Preferred Foot"] == "Left"].count()[0] #.count()[0] give total number
right = fifa.loc[fifa["Preferred Foot"] == "Right"].count()[0]
labels = ["Left", "Right"]
colors = ["red", "blue"]

plt.figure(figsize=(5,3), dpi=200)

plt.pie([left,right], labels = labels, colors = colors, autopct="%.2f %%") #%.2f meaning % in 2 decimals float. %% at the end shows % at the end of each values

plt.show()
```

# Pie Chart 2

```python
fifa["Weight"]
```

```python
fifa["Weight"] = [int(x.strip("lbs")) if type(x) == str else x for x in fifa["Weight"]] # Converting x to int, if x value is strm then remove lbs. else leave it as is.
fifa["Weight"]
```

```python
light = fifa.loc[fifa["Weight"]< 125].count()[0]
light_medium = fifa.loc[(fifa["Weight"] >= 125) & (fifa["Weight"] < 150)].count()[0]
medium = fifa.loc[(fifa["Weight"] >= 150) & (fifa["Weight"] < 175)].count()[0]
medium_heavy = fifa.loc[(fifa["Weight"] >= 175) & (fifa["Weight"] < 200)].count()[0]
heavy = fifa.loc[fifa["Weight"] >= 200].count()[0]

plt.style.use("ggplot") #applkying style

weight = [light, light_medium, medium, medium_heavy, heavy]
labels = ["Light", "Light Medium", "Medium", "Medium Heavy", "Heavy"]
explode = (.4,.1,0,0,.4)


plt.figure(figsize=(5,3), dpi=200)

plt.title("Weight Distribution of FIFA Players (in lbs)")

plt.pie(weight, labels = labels, autopct="%.2f %%", explode=explode) #%.2f meaning % in 2 decimals float. %% at the end shows % at the end of each values

plt.show()
```

```python
plt.style.use("default")

barcelona = fifa.loc[fifa["Club"] == "FC Barcelona"]["Overall"]
madrid = fifa.loc[fifa["Club"] == "Real Madrid"]["Overall"]
new_england = fifa.loc[fifa["Club"] == "New England Revolution"]["Overall"]
labels = ["Barcelona", "Madrid", "New England"]

plt.figure(figsize=(3,6), dpi=200)
plt.ylabel("Overall Rating")

boxes = plt.boxplot ([barcelona, madrid, new_england], labels = labels, patch_artist=True, medianprops = {"linewidth":2})

for box in boxes["boxes"]:
    # Set edge colour
    box.set(color="blue", linewidth = 2)
    
    # Change fill colour
    box.set(facecolor="black")
    
plt.show()
```
